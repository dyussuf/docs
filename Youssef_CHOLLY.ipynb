# Un Modèle de Battement du Coeur.

## le système d’équations différentielles :

$$x'(t)=\epsilon(y(t)-\frac{1}{3}x^3(t)+x(t))$$
$$y'(t)=-x(t)$$


Pour  $\epsilon=1$:

```python
from math import*
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
def predator_prey_system(t, y):
    return np.array([(y[1] - 1/3*y[0]**3+y[0]), -y[0]])
```

```python
x = np.linspace(-4, 4, 25)
y = np.linspace(-3, 3, 25)
X, Y = np.meshgrid(x, y)
dx, dy = predator_prey_system(1, np.array([X, Y]))
fig, ax = plt.subplots()
ax.quiver(X, Y, dx, dy)
ax.set_title("agramme des phases (e=1)")
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.show()
```
