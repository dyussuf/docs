Pour chaque étudiant, il faut déclarer les notes et récrire les instructions pour calculer et afficher leur moyenne; Ce n'est évidemment pas la bonne manière de programmer et faciliter la tâche ! Pour cela on recourt à ce qu'on appelle une ***fonction***.

En programmation, une fonction est un bloc d'instructions réutilisable; Au lieu de répéter les instructions, on les met sous forme d'une fonction et on l'appelle quand on en a besoin.

Un ensemble de fonctions regroupées est appelé ***bibliothèque*** (ou **module**). Par exemple la bibliothèque ***math*** de Python regroupe des fonctions mathématiques comme $cos$, $sin$, $sqrt$, . . . 

Jusqu'à présent, on n'a utilisé que des fonctions prédéfinies (c'est-à-dire des fonctions déjà définies quelque part). En effet, le langage Python fournit déjà un certain nombre de fonctions pour éviter de devoir les écrire à  la main. Dans ce cas, on a soit :

- utilisé directement ces fonctions car elles sont directement intégrées comme $range$, ... ;  
- importé ces fonctions depuis une bibliothèque, par exemple la fonction $sqrt$ du module ***math*** et $randint$ du module ***random***.

Une fonction comporte deux parties :  

1. **la définition**  
2. **l'appel**

# Définition  

La première étape pour pouvoir utiliser une fonction est de la définir; Ça consiste à :  

1. définir un **nom** pour la fonction ;  
2. définir les **entrées** qu'elle attend et les **sorties** qu'elle renvoie ;  
3. définir le cœur de sa logique.  

On peut voir la définition d'une fonction comme une opération analogue à une affectation, avec la différence qu'ici, nous n'allons pas stocker une simple valeur en mémoire, mais une séquence d'instructions.


## Syntaxe

La structure de la définition d'une fonction est la suivante :  

&nbsp;

**def** nom **(** paramètres **)** **:**

> bloc d'instructions (corps)

> [**return** expression]

&nbsp;

- **def**  
  Un mot-clé obligatoire indiquant qu'on définit (**def**ine) une fonction. 
  
&nbsp;

- **nom**  
  Un identifiant choisi pour la fonction, respectant les mêmes règles que pour les noms de variables. 
  
&nbsp;

- **paramètres**  
  Une séquence d'identifiants séparés par des virgules, indiquant quelles entrées la fonction attend. Les paramètres sont facultatifs, mais les parenthèses sont obligatoires. Si la fonction n’a pas besoin de paramètres, les parenthèses restent vides.  
  
&nbsp;

- **corps**  
  Un bloc d'instructions exécutées lorsque la fonction est appelée. Dans ce bloc, on peut utiliser les variables spécifiées comme entrées de la fonction. Pour marquer le début et la fin de la fonction, le bloc doit être décalé d'un niveau (indentation) par rapport à sa définition.
  
&nbsp;

- **`return expression`**  
  - La fonction renverra l'évaluation de l'expression comme sortie. Cette sortie peut être affichée directement sur l'écran ou affectée à une variable.
  - Une fonction peut avoir une ou plusieurs instructions ***return***, dans ce cas, l'exécution de la fonction se termine dès que l'une de ces instructions est atteinte.
  - Si aucune instruction ***return*** n'est présente, la fonction se termine lorsque son exécution atteint la dernière ligne, et la sortie sera automatiquement ***None*** (Rien).

## Exemple

On reprend l'exemple de la moyenne :

- La fonction peut être nommée ***moyenne***;
- les entrées  sont les trois notes;
- la sortie est leur moyenne.

```python
def moyenne(n1, n2, n3):

    m = round((n1+n2+n3)/3, 2)

    return m
```

# Appel  

C'est la partie où on utilise (appelle) une fonction définie; Il suffit d'écrire son nom suivi de parenthèses, entre lesquelles toutes les entrées (arguments) nécessaires, s'il y en a.

Les arguments représentent les valeurs réelles passées à la fonction lors de son appel. Par défaut, une fonction doit être appelée avec autant d'arguments qu'il y a de paramètres définis.

## Exemples

La fonction ***moyenne*** est définie avec trois paramètres, on doit toujours l'appeler avec **exactement** trois arguments, ni plus, ni moins.

```python
# Étudiant numéro 1

print(moyenne(13, 10, 17))
```

```python
# Étudiante numéro 2

x = moyenne(17, 16, 15)

print(x)
```

C'est la bonne façon d'appeler la fonction ***moyenne***, en passant trois entrées, sinon une erreur se produit :

```python
moyenne(17)
```

## Arguments par défaut

Parfois, dans la définition d'une fonction, on a besoin de spécifier une valeur par défaut pour un ou plusieurs paramètres. Cela peut être fait en indiquant la valeur par défaut directement dans la définition, comme ceci :

```python
def moyenne(n1, n2, n3=18):

    m = round((n1+n2+n3)/3, 2)

    return m

# le paramètre n3 a 18 comme valeur par défaut, donc on peux passer 2 arguments au lieu de 3 :

moyenne(10, 13)
```

```python
# ou trois arguments si on dispose de la valeur de la troisième note :

moyenne(17, 16, 15)
```

# Portée (*Scope*)  

Chaque objet possède sa propre portée qui peut être vue comme un environnement dans lequel un objet existe et qui détermine avec quels autres objets il peut interagir.

En général, tout ce qui est défini au niveau supérieur d'un programme Python fait partie d'un environnement unique appelé **global**, et tout ce qui est défini à l'intérieur d'une fonction appartient à un environnement distinct, appelé environnement **local**.  


## Règles  

1. **Un objet global est accessible partout** :  
   Ce qui est défini au niveau global est accessible depuis n'importe quel environnement, qu'il soit global ou local.  

2. **Un objet local est restreint à son propre environnement** :  
   Ce qui est défini à l'intérieur d'une fonction n'est accessible que dans cette fonction, mais pas en dehors de la fonction.   

```python
x1 = 1000

def test():
    x2 = 77
    print("Environment local (à l'intérieur d'une fonction) :")
    print("variable globale : ", x1)   
    print("variable locale : ", x2)   

test()
```

Dans ce cas la variable $x1$ est de portée globale, alors on peut l'utiliser localement.

```python
print("Environment global (à l'extérieur) :")
print("variable globale : ", x1)
print("variable locale : ", x2)    
```

Par contre, $x2$ est de portée locale, alors on ne peut pas l'utiliser globalement, d'où l'erreur.

# lambda

lambda est une fonction anonyme (sans nom, sauf si elle est assignée à une variable) qu'on definie en une seule ligne. Elle est utile pour définir une fonction simple et temporairement utilisée.


## Syntaxe

**lambda** paramètres : expression

Après le mot clé **lambda** on écrit les *paramètres* d'entrée de la fonction, séparés par des virgules, et une seule *expression* après **:** qui est évaluée et dont le résultat est renvoyé.

## Exemples

```python
# addition de deux nombres

addition = lambda x1, x2 : x1 + x2

addition(3, 5)
```

```python
# test de parité 

pair = lambda x : x%2 == 0

pair(7)
```

Une lambda, comme on le verra ultérieurement, est couramment définie à l'endroit où elle est nécessaire, sans l'assigner à une variable, pour des opérations rapides dans des contextes temporaires.

# Assertion

En Python, l'instruction ***assert*** sert à s'assurer qu'une condition est remplie pour que le programme fonctionne comme prévu.

## Syntaxe

**assert**   condition[,   message]

- **condition** : Une expression booléenne qui doit être évaluée à $True$ pour que l'exécution continue.
- **message** : (optionnel) qui sera affiché si l'expression est évaluée à $False$.

## Exemple

```python
x = eval(input("Donner un nombre non nul : "))

assert x != 0, "Le nombre doit être différent de 0"
```

Dans cet exemple, on demande à l'utilisateur un nombre non nul, si le nombre donné est nul, 
une erreur sera affichée avec le message ***Le nombre doit être différent de 0***.
